## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
ds = pd.read_csv('Task (2) Dataset.csv')
```

```python
ds.describe()
```

```python
ds.head()
```

### Identify and analyze patterns in the data.

```python
ds.info()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
# Bar Chart for average insurance premium by state
plt.figure(figsize=(10, 6))
sns.barplot(ds, x='City', y='JamsCount', palette='muted')
plt.title('Numbers of JamsCount based on citys')
plt.xlabel('City')
plt.ylabel('JamsCount')
plt.xticks(rotation=90)# for the label button the chart how its show 
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
sns.scatterplot(ds, x='JamsLength', y='JamsDelay')
```

```python
dataset = pd.read_csv('Task (1) Dataset.csv')
```

```python
column_to_merge = dataset[['Weather_conditions']]
```

```python
ds = pd.concat([ds,column_to_merge], axis=1)
```

```python
ds.head()
```

```python
sns.scatterplot(ds, x='Weather_conditions', y='TrafficIndexLive')
```
